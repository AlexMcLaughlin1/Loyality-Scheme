```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
lognormal_values_long_tail = np.random.lognormal(mean=1, sigma=0.4, size=10000) - 1
lognormal_values_long_tail = np.maximum(lognormal_values_long_tail, 1)
# Adjust log-normal parameters for a longer tail

# Plot the modified log-normal distribution
plt.figure(figsize=(10, 5))
plt.hist(lognormal_values_long_tail, bins=100, density=True, alpha=0.6, color='orange', label="Log-Normal (mean=3, sigma=1.2)")

# Formatting
plt.xlabel("Order Value")
plt.ylabel("Density")
plt.title("Log-Normal Distribution with a Longer Tail")
plt.legend()

# Show plot
plt.show()
```
